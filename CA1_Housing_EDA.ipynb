# ***Data Visualization***

# ***Data Preprocessing***

```python
#Copy the original dataset For this Section
house_df_preprocess = house_df.copy()
```

```python
print(house_df_preprocess.shape)
print(house_df_preprocess.dtypes)
```

```python
house_df_preprocess.isnull().sum()
```

```python
#Percentage of null values to total records (Help determine which records should stay or be deleted)
null_house = house_df_preprocess.isnull().mean().round(4)

#Used Chat to find create a lambda code that helps highlight features with null values
#This will help identify features with missing values to decide what to do to fill them
#https://chatgpt.com/share/697e84bc-70e0-8001-8078-887e76ec9efd
null_house.to_frame('Null %').style.applymap(
    lambda v: 'background-color: salmon' if v > 0.0 else ''
)
```

```python
#LotFrontage Fill NA for the 17% missing Values
house_df_preprocess['LotFrontage'].fillna(house_df_preprocess['LotFrontage'].median(), inplace=True)
```

```python
#MassVnrArea Fill NA for the 0.5% missing Values
house_df_preprocess['MasVnrArea'].fillna(house_df_preprocess['MasVnrArea'].median(), inplace=True)
```

```python
#GarageYrBuilt Fill NA for the 5.6% missing Values
house_df_preprocess['GarageYrBlt'].fillna(house_df_preprocess['GarageYrBlt'].median(), inplace=True)
```

```python
#Breakdown of missing values after filling the previous columns
print(house_df_preprocess.isnull().sum())
```

```python
#Dropping Columns: Alley, PoolQC, Fence, amd MiscFeature due to over 70% missing null-values
#Will also drop MasVnrType and FireplaceQu since around 50% of values are missing and feel miniscule to housing price
house_df_preprocess_fixed = house_df_preprocess.drop(columns=['Alley','PoolQC','Fence','MiscFeature','MasVnrType','FireplaceQu'])

print(house_df_preprocess_fixed.dtypes)
#house_df_preprocess_fixed will be used for the rest of this section
```

```python
#For the 10 object columns with null values replaced with Unknown to preserve columns and not erase to many records
house_df_preprocess_fixed.fillna("Unknown", inplace=True)

#all columns now have zero null values for analysis
print(house_df_preprocess_fixed.isnull().sum())
```

```python
#Summary Stats for Numeric Column
print(house_df_preprocess_fixed["SalePrice"].describe())
```

```python
#Skewness and Kurtosis caculations of Sale Price

#Sale Price is highly positively skewed with a score of 1.88
print(f"Sale Price skew:",house_df_preprocess_fixed['SalePrice'].skew())

#With a score of 6.54 Sale Price has an excessive amount outliers
print(f"Sale Price kurtosis:",house_df_preprocess_fixed['SalePrice'].kurtosis())
```

```python
#Correlation of two numeric columns
print(f"Correlation of Sale Price to LotArea:", house_df_preprocess_fixed['SalePrice'].corr(house_df_preprocess_fixed['LotArea']))
#0.264 correlation shows a minimal positive correlation between Sale Price and Lot Area
```

```python
#t-test to compare means of numeric columns
from scipy.stats import ttest_ind

t, p = ttest_ind(house_df_preprocess_fixed['SalePrice'], house_df_preprocess_fixed['LotArea'])
print(t,p)
```

```python
#ANOVA test comparing the means of Two numeric columns
from scipy.stats import f_oneway

f, p = f_oneway(house_df_preprocess_fixed['SalePrice'], house_df_preprocess_fixed['LotArea'])
print(f, p)
```

```python
#Drop Non-Predictive Columns
#Deleted ID since it was the only with no Predictive power (other columns may provide insight)
house_df_preprocess_fixed = house_df_preprocess_fixed.drop(columns=['Id'])

print(house_df_preprocess_fixed.dtypes)
```

```python
#encoding Categorical Variables to Numeric Columns
coded_house_df = house_df_preprocess_fixed

print(coded_house_df.dtypes)
```

```python
#Label Encoding for Columns with a Ordinal Ranking (Happened to share a similar ranking system)
#made with assistance from Chat to ensure Ordinal scale and function works properly
#https://chatgpt.com/share/697ff552-b600-8001-a350-4371251900d1
quality_order = ['Unknown','Po','Fa','TA','Gd','Ex']

cols = [
    'ExterQual','ExterCond','BsmtQual','BsmtCond',
    'HeatingQC','KitchenQual','GarageQual','GarageCond'
]

coded_house_df[cols] = (
    coded_house_df[cols]
    .fillna('Unknown')
    .apply(lambda s: pd.Categorical(s, categories=quality_order, ordered=True).codes)
)
```

```python
#One Hot Encoding for Non-Ordinal Columns
coded_house_df = pd.get_dummies(coded_house_df,columns=['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
'Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','Foundation','BsmtExposure','BsmtFinType1',
'BsmtFinType2','Heating','CentralAir','Electrical','Functional','GarageType','GarageFinish','PavedDrive','SaleType','SaleCondition'])
```
